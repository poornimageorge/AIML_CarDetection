## Reducing Input Data

```python
import random
```

```python
trainMergedImageDF_reduced = trainMergedImageDF[trainMergedImageDF['Imageclass'].isin([10,40,80,120,160,180])]
```

```python
trainMergedImageDF_reduced
```

```python
#Use Label encoder available in Scikit Learn
from sklearn import preprocessing

#Label Encode class and add a 'label' column to the dataframe
le = preprocessing.LabelEncoder()
trainMergedImageDF_reduced['label'] = le.fit_transform(trainMergedImageDF_reduced['ImageClassName'])
trainMergedImageDF_reduced.head()
```

```python
#Object detection API expects index to start from 1 (and not 0)
trainMergedImageDF_reduced['label'] = trainMergedImageDF_reduced['label'] + 1
trainMergedImageDF_reduced.label.unique()
```

```python
#Create a dictionary of Label and Class. This will be useful for building our second input to Model training
label_class_dict = dict(zip(trainMergedImageDF_reduced['label'], trainMergedImageDF_reduced['ImageClassName']))
```

```python
print(label_class_dict)
```

```python
testMergedImageDF_reduced = testMergedImageDF[testMergedImageDF['Imageclass'].isin([10,40,80,120,160,180])]
testMergedImageDF_reduced
```

```python
testMergedImageDF_reduced.Imageclass.unique()
```

```python
def fn(df):
    if df['Imageclass'] == 10:
        return 1
    elif df['Imageclass'] == 40:
        return 2
    elif df['Imageclass'] == 80:
        return 3
    elif df['Imageclass'] == 120:
        return 4
    elif df['Imageclass'] == 160:
        return 5
    elif df['Imageclass'] == 180:
        return 6
```

```python
testMergedImageDF_reduced['label'] = testMergedImageDF_reduced.apply(lambda x: fn(x), axis=1)
```

## Generate label_map

```python

#Build a pbtxt label file using label and class name
#This is required by Object detection API
#You can prepare it manually as well.

pbtxt_file_txt = ''
for label in sorted(label_class_dict.keys()):
    
    pbtxt_file_txt += "item {\n  id: " + str(label) + "\n  name: '" +  label_class_dict[label] + "'\n}\n\n"

with open('label_map.txt','w') as pbfile:
    pbfile.write(pbtxt_file_txt)
```
